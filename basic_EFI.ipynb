```python
from viresclient import SwarmRequest
import datetime as dt

request = SwarmRequest("https://staging.viresdisc.vires.services/openows")

request.set_collection("SW_OPER_EFIA_PL_1B")

request.set_products(measurements=request.available_measurements("EFI"))
```

```python
# No data until 2014,1,7
# How to handle these errors?
start_time = dt.datetime(2014,1,1)
end_time = dt.datetime(2014,1,2)
data = request.get_between(start_time, end_time, asynchronous=False)
```
